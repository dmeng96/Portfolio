# 1.3 - Regression Linéaire

```python
def detection_billets(fichier_csv):

    # Tranformation 'is_genuine' avec les TRUE = 0
    # Puisqu'il s'agit de détecter si un billet est faux il faut chager le nom de la colonne is_genuine
    data_final = data.rename(columns={'is_genuine': 'is_false'})
    data_final['is_false'] = np.where(data_final['is_false'] == True, 0, 1)

    X_test = data.drop(columns=['is_genuine'])
    pipe = joblib.load('logistic-regression-gs.joblib')
    y_pred = pipe.predict(X_test)
    pred = pd.DataFrame(y_pred)
    pred = pred.rename(columns={0: 'Prédiction'})
    pred = pred.replace({0: 'Vrai billet', 1: 'Faux billet'})

    y_proba = pipe.predict_proba(X_test)
    proba = pd.DataFrame(y_proba)
    proba = proba.rename(columns={0: 'Probabilité de faux', 1: 'Probabilité de vrais'})
    
    # Concaténation des données
    prediction = pd.concat([pred, proba], axis = 1)
    final = pd.concat([prediction, data], axis = 1)    
    return(final)
```
